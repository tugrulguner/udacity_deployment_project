```python
# Import libraries
import pandas as pd
```

```python
# Read the raw data
data = pd.read_csv('data/census.csv')
data.head(20)
```

```python
# Check if there is any nan values
data.isna().sum()
```

```python
# Remove the empty spaces from the column names
column_names = [col.replace(' ', '') for col in data.columns]
data.set_axis(column_names, axis=1, inplace=True)
data
```

```python
# Remove the empty spaces from all strings in the data
data.replace(' ', '', regex=True, inplace=True)
```

```python
data.describe()
```

```python
data.to_csv('/data/cleaned_data.csv')
```
